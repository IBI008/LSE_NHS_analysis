```python
# Create a new DataFrame consisting of the month of appointment and the number of appointments.
nc_month_spread = nc.groupby(['appointment_month','service_setting'])['count_of_appointments'].sum().reset_index()

# View the DataFrame.
nc_month_spread
```

```python
# Create a boxplot to investigate spread of service settings.
nc_month_spread_boxplot = sns.boxplot(data=nc_month_spread, x='service_setting', y='count_of_appointments')

# Set labels and title.
nc_month_spread_boxplot.set_xlabel("Service setting")
nc_month_spread_boxplot.set_ylabel("Sum of appointments")
nc_month_spread_boxplot.set_title("Spread of service settings")
```

```python
# Create a subset without GP.
nc_month_spread_no_GP = nc_month_spread[nc_month_spread['service_setting'] != 'General Practice']

# View the DataFrame.
nc_month_spread_no_GP.head()
```

```python
# Create a boxplot to investigate the service settings without GP.
nc_month_spread_boxplot_no_GP = sns.boxplot(data=nc_month_spread_no_GP, x='service_setting', \
                                      y='count_of_appointments')

# Set labels and title.
nc_month_spread_boxplot_no_GP.set_xlabel("Service setting")
nc_month_spread_boxplot_no_GP.set_ylabel("Sum of appointments")
nc_month_spread_boxplot_no_GP.set_title("Spread of service settings excluding GP")
```
